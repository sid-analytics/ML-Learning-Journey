```python
newdf = pd.DataFrame(np.random.rand(30,5), index = np.arange(30))
newdf
```

```python
newdf2 = newdf               # It has became view means they point to same memory location
newdf2[0][0] = 32
newdf2
```

```python
newdf
```

```python
newdf3 = newdf.copy()                 # newdf.copy() creates the copy of the newdf
newdf3[0][2] = 1
newdf3
```

#### The given error when we give a value to a particular index gives error as sometimes pandas allocate it as view and sometimes as copy so it is very inefficient to change dataframe or series like that so instead we use loc to modify or update the values

```python
newdf
```

```python
newdf.columns = list("ABCDE")
newdf
```

```python
newdf.loc[0, 0] = 45
newdf
```

```python
newdf.drop(0, axis = 1)
```

```python
newdf                                    # It doesn't change until you update the df
```

```python
newdf = newdf.drop(0)
newdf
```

```python
# To observe certain sections only
newdf.loc[[2, 3], ["D", "E"]]
```

```python
newdf.loc[(newdf["E"] < 0.3)]
```

```python
# To do indexing from index only then use iloc
newdf.iloc[0, 3]
```

```python
newdf.drop([5], inplace = True)               # This replaces original dataframe
```
