### Numpy Operations

```python
#slicing
arr = np.arange(5, 50, 5)
print(arr[4:7])
```

```python
rand = np.random.random((3,3))
print(rand)
print("Specific Element", rand[2,2])
print("Entire row", rand[0])
print("Entire column", rand[:, 2])
```

```python
unsorted = np.array([2,7,1,5,0])
print(np.sort(unsorted))

unsort = np.array([[4, 6],
                   [7, 2],
                   [1, 4]])
print("Whole sort", np.sort(unsort, axis = 0))
print("Sort in each row", np.sort(unsort, axis = 1))
```

```python
# Expressions can be used instead of indexes in numpy
# called as filtering and expression can be called as mask

numbers = np.array([1, 2, 3, 4, 5, 6, 7, 8])
even = numbers[numbers % 2 == 0]
print(even)

mask = numbers < 7
print(numbers[mask])
```

```python
# fancy indexing

indices = [3, 4, 1]
print(numbers[indices])
```

```python
# where in numpy

where_result = np.where(numbers>5)
print(where_result)
print(numbers[where_result])
```

```python
arr1 = np.array([2, 4, 5])
arr2 = np.array([2, 7, 9])
print(arr1 - arr2)
```

```python
print(np.concatenate((arr1, arr2)))
```

```python
# Combining arrays
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([5, 6])
arr3 = np.array([[7], [8]])

# vertical stacking
print(np.vstack((arr1, arr2)))

# horizontal stacking
print(np.hstack((arr1, arr3)))
```

```python
arr = np.array([1, 2, 3, 4, 5])
deleted1 = np.delete(arr, 0)
deleted2 = np.delete(arr, [0, 4, 2])              #index of start, end, step
print(deleted1, "\n", deleted2)
```
